```python
import tensorflow as tf
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing.text import one_hot #allows for OHE
```

```python
#sentences

sent=[  'the glass of coke',
     'the glass of juice',
     'the cup of tea',
    'He is a good boy',
     'I am a good programmer',
     'understand the meaning of life',
     'the girls are good']
```

```python
sent
```

```python
#vocab
vocab_size=10000
```

```python
onehot_ind=[one_hot(words,vocab_size)for words in sent]
print(onehot_ind)   #Catching the indexes from OHE
```

```python
#All my sentences should have the same dimensions - post and pre padding
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## pre padding
max_sentl=8
emb_docs=pad_sequences(onehot_ind,padding='pre',maxlen=max_sentl)
print(emb_docs)
```

```python
#feature representation dimension = 10
dim = 10
```

```python
model = Sequential()
model.add(Embedding(vocab_size,10,input_length=max_sentl))
model.compile('adam','mse')
```

```python
model.summary()
```
