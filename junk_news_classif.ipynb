```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df = df.dropna()  #Since it is text data
```

```python
df.shape, df.isnull().sum()
```

```python
X = df.drop('label',axis=1)
```

```python
y= df['label']  #op
```

```python
X.shape, y.shape
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
vocab_size=10000
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
msg = X.copy()
```

```python
msg
```

```python
msg.reset_index(inplace=True)
```

```python
msg
```

```python
import nltk
import re  #regular expressions
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer ##stemming purpose
ps = PorterStemmer()
corpus = []
for i in range(0, len(msg)):
    review = re.sub('[^a-zA-Z]', ' ', msg['title'][i])  #replace all the other special charac with blank space
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus[100]
```

```python
ohe = [one_hot(words,vocab_size) for words in corpus]
```

```python
ohe
```

```python
corpus[200], ohe[200]
```

```python
max_sentl = 18
```

```python
emdb_docs = pad_sequences(ohe,padding='post',maxlen=max_sentl)
emdb_docs
```

```python
emdb_docs[300]
```

```python
#Model
embedding_vector_features=50 #features representation
model=Sequential()
model.add(Embedding(vocab_size,embedding_vector_features,input_length= 18))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))  #o/p is binary
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(emdb_docs), y.shape
```

```python
import numpy as np
```

```python
x_final = np.array(emdb_docs)
y_final = np.array(y)
```

```python
x_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_final, y_final, test_size=0.33, random_state=20)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=8,batch_size=32)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.5, 1,0) ##AUC ROC Curve
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
